```python
#Task 3: Documenting Cumulative Score
# Install necessary library for exporting to Excel
!pip install openpyxl

# Import required libraries
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the dataset
df = pd.read_csv("../Projects/GrennTree/test.csv")

# 2. Combine Subject and Body into a single 'full_text' column
df["full_text"] = df["Subject"].fillna('') + " " + df["body"].fillna('')

# 3. Initialize sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# 4. Function to assign sentiment label and score based on compound score
def analyze_sentiment(text):
    compound = analyzer.polarity_scores(text)["compound"]
    if compound >= 0.05:
        return "Positive", 1  # Positive sentiment
    elif compound <= -0.05:
        return "Negative", -1  # Negative sentiment
    else:
        return "Neutral", 0  # Neutral sentiment

# 5. Apply sentiment function to analyze sentiment for each row
df[["Sentiment", "score"]] = df["full_text"].apply(lambda x: pd.Series(analyze_sentiment(x)))

# 6. Convert 'date' to datetime and extract Month-Year for time-based analysis
df["date"] = pd.to_datetime(df["date"])
df["Month"] = df["date"].dt.to_period("M").astype(str)

# 7. Rename 'from' column to 'Employee' for better clarity
df.rename(columns={"from": "Employee"}, inplace=True)

# 8. Group by Month and Employee to calculate monthly sentiment scores
monthly_scores = df.groupby(["Month", "Employee"])["score"].sum().reset_index()
monthly_scores.rename(columns={"score": "Monthly_Score"}, inplace=True)

# Function to get top employees for each month based on sentiment
def get_top_employees_for_months(monthly_scores, sentiment_type='positive'):
    top_employees_data = []  # List to store top employees data for CSV export
    months = monthly_scores["Month"].unique()  # Get all unique months

    for month in months:
        monthly_data = monthly_scores[monthly_scores["Month"] == month]

        # Identify top positive employees
        if sentiment_type == 'positive':
            top_employees = (
                monthly_data[monthly_data["Monthly_Score"] > 0]
                .sort_values(["Monthly_Score"], ascending=False)  # Sort to get highest positive scores
                .head(3)  # Get top 3
            )
            sentiment_label = 'Positive'
        else:
            # For negative sentiment, get the most negative scores
            top_employees = (
                monthly_data[monthly_data["Monthly_Score"] < 0]
                .sort_values(["Monthly_Score"], ascending=True)  # Sort from most negative
                .head(3)  # Get top 3
            )
            sentiment_label = 'Negative'

        # Append top employees' data for the current month
        for _, row in top_employees.iterrows():
            top_employees_data.append({
                "Month": row["Month"],
                "Employee": row["Employee"],
                "Sentiment": sentiment_label,
                "Score": row["Monthly_Score"]
            })
    
    # Create a DataFrame with the collected data
    top_employees_df = pd.DataFrame(top_employees_data)
    return top_employees_df

# 1. Get Top Positive Employees
top_positive_employees = get_top_employees_for_months(monthly_scores, sentiment_type='positive')

# 2. Get Top Negative Employees
top_negative_employees = get_top_employees_for_months(monthly_scores, sentiment_type='negative')

# Combine both positive and negative employee results
final_result = pd.concat([top_positive_employees, top_negative_employees])

# 9. Save the result to an Excel file
final_result.to_excel("top_positive_negative_employees.xlsx", index=False)

# Display the first few rows of the result
final_result.head(76)
```

  - The **`get_top_employees_for_months`** function identifies the **top 3 employees** for each month based on their **sentiment score**.

```python
#Task 4: Employee Ranking

# Import required libraries
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the dataset
df = pd.read_csv("../Projects/GrennTree/test.csv")

# 2. Combine Subject and Body into a single 'full_text' column
df["full_text"] = df["Subject"].fillna('') + " " + df["body"].fillna('')

# 3. Initialize sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# 4. Function to assign sentiment label and score based on compound score
def analyze_sentiment(text):
    compound = analyzer.polarity_scores(text)["compound"]
    if compound >= 0.05:
        return "Positive", 1  # Positive sentiment
    elif compound <= -0.05:
        return "Negative", -1  # Negative sentiment
    else:
        return "Neutral", 0  # Neutral sentiment

# 5. Apply sentiment function to analyze sentiment for each row
df[["Sentiment", "score"]] = df["full_text"].apply(lambda x: pd.Series(analyze_sentiment(x)))

# 6. Convert 'date' to datetime and extract Month-Year for time-based analysis
df["date"] = pd.to_datetime(df["date"])
df["Month"] = df["date"].dt.to_period("M").astype(str)

# 7. Rename 'from' column to 'Employee' for better clarity
df.rename(columns={"from": "Employee"}, inplace=True)

# 8. Group by Month and Employee to calculate monthly sentiment scores
monthly_scores = df.groupby(["Month", "Employee"])["score"].sum().reset_index()
monthly_scores.rename(columns={"score": "Monthly_Score"}, inplace=True)

# ---------------------- VISUALIZATIONS ----------------------

# Function to plot the top 3 positive/negative employees for each month
def plot_top_employees_for_months(monthly_scores, sentiment_type='positive'):
    # Unique months from the dataset
    months = monthly_scores["Month"].unique()
    
    # Loop through each month
    for month in months:
        monthly_data = monthly_scores[monthly_scores["Month"] == month]
        
        if sentiment_type == 'positive':
            top_employees = (
                monthly_data[monthly_data["Monthly_Score"] > 0]
                .sort_values(["Monthly_Score"], ascending=False)
                .head(3)
            )
            title = f"Top 3 Positive Employees in {month}"
            palette = "viridis"
        else:
            # Improved: We will only consider the top 3 most negative sentiment scores (most extreme negative values)
            top_employees = (
                monthly_data[monthly_data["Monthly_Score"] < 0]
                .sort_values(["Monthly_Score"], ascending=True)  # Sort from most negative to least negative
                .head(3)  # Top 3 negative employees
            )
            title = f"Top 3 Negative Employees in {month}"
            palette = "rocket"

        # Plotting each month's top employees if there are any data points
        if not top_employees.empty:
            plt.figure(figsize=(6, 4))
            sns.barplot(
                data=top_employees,
                x="Monthly_Score", y="Employee",
                palette=palette
            )
            plt.title(title)
            plt.xlabel("Monthly Sentiment Score")
            plt.ylabel("Employee")
            plt.tight_layout()
            plt.show()

# 1. Plot Top Positive Employees for each month
plot_top_employees_for_months(monthly_scores, sentiment_type='positive')

# 2. Plot Top Negative Employees for each month
plot_top_employees_for_months(monthly_scores, sentiment_type='negative')
```

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
import seaborn as sns

# ----------------- Step 1: Load and Prepare Data ----------------- #
# Load the dataset
df = pd.read_csv("../Projects/GrennTree/test.csv")

# Combine Subject and Body text into one column 'full_text' for sentiment analysis
# This ensures both Subject and Body content are considered for sentiment scoring
df["full_text"] = df["Subject"].fillna('') + " " + df["body"].fillna('')

# Initialize the SentimentIntensityAnalyzer from VADER
analyzer = SentimentIntensityAnalyzer()

# Sentiment analysis function
# This function returns sentiment label and score based on the 'compound' score from VADER
def analyze_sentiment(text):
    compound = analyzer.polarity_scores(text)["compound"]
    if compound >= 0.05:
        return "Positive", 1    # Sentiment is Positive
    elif compound <= -0.05:
        return "Negative", -1   # Sentiment is Negative
    else:
        return "Neutral", 0     # Sentiment is Neutral

# Apply the sentiment analysis to the 'full_text' column
df[["Sentiment", "score"]] = df["full_text"].apply(lambda x: pd.Series(analyze_sentiment(x)))

# ----------------- Observations on Step 1 ----------------- #
# - Combining the Subject and Body ensures we analyze the entire email text.
# - The analyzer checks polarity using VADER, which returns a compound score.
# - Depending on the compound score, the text is classified as Positive, Negative, or Neutral.

# ----------------- Step 2: Data Preprocessing ----------------- #
# Convert the 'date' column to datetime format for easier extraction of Month-Year
df["date"] = pd.to_datetime(df["date"], errors='coerce')  # Handling any invalid date formats gracefully
df = df.dropna(subset=["date"])  # Dropping rows where 'date' is missing after conversion

# Extract Month and Year in Period format (e.g., '2025-05')
df["Month"] = df["date"].dt.to_period("M").astype(str)

# Rename the 'from' column to 'Employee' for clarity
df.rename(columns={"from": "Employee"}, inplace=True)

# ----------------- Observations on Step 2 ----------------- #
# - `errors='coerce'` will convert any invalid date entries to NaT (Not a Time), and we drop them.
# - Extracting Month-Year helps us group by month for the ranking.
# - Renaming 'from' to 'Employee' makes it clearer that this represents the sender.

# ----------------- Step 3: Calculate Monthly Scores ----------------- #
# Group by 'Month' and 'Employee' to calculate sentiment scores for each employee each month
monthly_scores = df.groupby(["Month", "Employee"])["score"].sum().reset_index()

# Rename the 'score' column to 'Monthly_Score' for clarity
monthly_scores.rename(columns={"score": "Monthly_Score"}, inplace=True)

# ----------------- Observations on Step 3 ----------------- #
# - Grouping by 'Month' and 'Employee' ensures that we calculate total sentiment scores per month per employee.
# - Summing the scores gives us the overall sentiment for each employee in each month.

# ----------------- Step 4: Ranking Top 3 Positive ----------------- #
# Filter for employees with positive scores
top3_positive = (
    monthly_scores[monthly_scores["Monthly_Score"] > 0]
    .sort_values(["Month", "Monthly_Score", "Employee"], ascending=[True, False, True])  # Sorting for ranking
    .groupby("Month")
    .head(3)  # Select top 3 employees for each month
)

# Rank the employees within each month based on their Monthly_Score
top3_positive["Rank"] = top3_positive.groupby("Month")["Monthly_Score"] \
                                     .rank(method="first", ascending=False).astype(int)

# Assign 'Top Positive' as the category for these employees
top3_positive["Category"] = "Top Positive"

# ----------------- Observations on Step 4 ----------------- #
# - Sorting by 'Monthly_Score' in descending order ensures the highest scoring employees are ranked top.
# - Using `.head(3)` ensures that only the top 3 employees for each month are selected.

# ----------------- Step 5: Ranking Top 3 Negative ----------------- #
# Filter for employees with negative scores
top3_negative = (
    monthly_scores[monthly_scores["Monthly_Score"] < 0]
    .sort_values(["Month", "Monthly_Score", "Employee"], ascending=[True, True, True])  # Sorting for ranking
    .groupby("Month")
    .head(3)  # Select top 3 negative employees for each month
)

# Rank the employees within each month based on their Monthly_Score
top3_negative["Rank"] = top3_negative.groupby("Month")["Monthly_Score"] \
                                     .rank(method="first", ascending=True).astype(int)

# Assign 'Top Negative' as the category for these employees
top3_negative["Category"] = "Top Negative"

# ----------------- Observations on Step 5 ----------------- #
# - Sorting by 'Monthly_Score' in ascending order ensures that the most negative (lowest) scores are ranked top.
# - The `.head(3)` ensures that only the top 3 negative employees are selected each month.

# ----------------- Step 6: Combine Results ----------------- #
# Combine top positive and negative rankings
final_result = pd.concat([top3_positive, top3_negative], ignore_index=True)

# Reorganize the final output columns
final_result = final_result[["Month", "Employee", "Monthly_Score", "Rank", "Category"]]

# Export the final result to an Excel file for easy sharing and reporting
final_result.to_excel("top_employee_rankings_by_month.xlsx", index=False)

# Confirmation message
print(" Excel file generated: top_employee_rankings_by_month.xlsx")

# Display the first 10 rows of the final result for review
final_result.head(10)

# ----------------- Observations on Step 6 ----------------- #
# - Combining both the positive and negative rankings into one DataFrame allows easy access to all data.
# - Exporting the final results to Excel allows for further analysis or reporting.

```

```python
#Task 5: Flight Risk Identification

# Import necessary libraries
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from datetime import timedelta

# ------------------- Step 1: Load and Prepare Data ------------------- #
# Load the dataset
df = pd.read_csv("../Projects/GrennTree/test.csv")

# Combine 'Subject' and 'body' columns into a single 'full_text' column
df["full_text"] = df["Subject"].fillna("") + " " + df["body"].fillna("")

# Rename 'from' column to 'Employee' for clarity
df.rename(columns={"from": "Employee"}, inplace=True)

# Convert the 'date' column to datetime format and drop rows with invalid dates
df["date"] = pd.to_datetime(df["date"], errors='coerce')
df = df.dropna(subset=["date"])

# ------------------- Step 2: Perform Sentiment Analysis ------------------- #
# Initialize the SentimentIntensityAnalyzer for sentiment analysis
analyzer = SentimentIntensityAnalyzer()

# Define a function to detect sentiment based on VADER compound score
def detect_sentiment(text):
    compound = analyzer.polarity_scores(text)["compound"]
    if compound <= -0.05:
        return "Negative"  # Negative sentiment
    elif compound >= 0.05:
        return "Positive"  # Positive sentiment
    else:
        return "Neutral"  # Neutral sentiment

# Apply sentiment detection to the 'full_text' column
df["Sentiment"] = df["full_text"].apply(detect_sentiment)

# ------------------- Step 3: Filter Only Negative Emails ------------------- #
# Filter rows where the sentiment is 'Negative'
neg_df = df[df["Sentiment"] == "Negative"].copy()

# Select only relevant columns and sort by 'Employee' and 'date'
neg_df = neg_df[["Employee", "date"]].sort_values(by=["Employee", "date"])

# ------------------- Step 4: Flight Risk Detection ------------------- #
# Initialize an empty list to store flight risk data
flight_risks = []

# Loop through each employee's group of negative emails
for employee, group in neg_df.groupby("Employee"):
    dates = group["date"].tolist()  # Get list of negative email dates for the employee
    for i in range(len(dates)):
        count = 1  # Start counting from the first negative email
        window_start = dates[i]  # Set the start of the 30-day window
        for j in range(i + 1, len(dates)):
            # Check if the email is within the 30-day window
            if dates[j] - window_start <= timedelta(days=30):
                count += 1  # Increase the count for each additional negative email in the window
            else:
                break  # Exit the inner loop if the 30-day window is exceeded
        # If 4 or more negative emails are found within the 30-day window, flag as flight risk
        if count >= 4:
            flight_risks.append({
                "Employee": employee,
                "Start_Date": window_start.date(),  # Start date of the 30-day window
                "End_Date": (window_start + timedelta(days=30)).date(),  # End date of the 30-day window
                "Negative_Email_Count": count  # Total negative emails in the window
            })
            break  # Stop after first flight risk detected for this employee

# ------------------- Step 5: Output to Excel ------------------- #
# Convert the flight risks list to a DataFrame
flight_risk_df = pd.DataFrame(flight_risks)

# Save the flight risk DataFrame to an Excel file
flight_risk_df.to_excel("flight_risk_employees.xlsx", index=False)

# Print success message
print(" Flight risk report saved as 'flight_risk_employees.xlsx'")

# Display the first 10 rows of the flight risk report
flight_risk_df.head(10)

```

### Observations and Notes on the Code:

#### **Step 1: Data Loading & Preparation**
- **Loading**: Data is loaded from a CSV into a DataFrame (`df`).
- **Text Combination**: `Subject` and `Body` columns are merged into `full_text`.
- **Column Renaming**: `from` is renamed to `Employee` for consistency.
- **Date Conversion**: `date` column is converted to datetime format. Invalid dates are dropped.

#### **Step 2: Sentiment Analysis**
- **Sentiment Classification**: Using VADER sentiment analysis, text is classified as `Negative`, `Neutral`, or `Positive` based on the `compound` score.

#### **Step 3: Filtering Negative Sentiments**
- **Negative Emails**: A new DataFrame (`neg_df`) is created to filter only `Negative` sentiment emails.
- **Sorting**: The `neg_df` is sorted by `Employee` and `date`.

#### **Step 4: Flight Risk Detection**
- **Rolling Window Analysis**: For each employee, a sliding window of 30 days is used to count negative emails. If 4 or more negative emails are found within 30 days, the employee is flagged as a `flight risk`.
- **Break Condition**: The loop stops checking once the first `flight risk` is detected for each employee.

#### **Step 5: Output**
- **Export**: The flight risk data is exported to an Excel file (`flight_risk_employees.xlsx`).

#### **Additional Notes**
- **Employee Risk**: Employees with 4 or more negative emails within 30 days are flagged for potential flight risk.
- **Efficient Filtering**: The use of a rolling window optimizes the detection of recurring negative sentiment.


```python
# Task 6: Predictive Modeling
# Import necessary libraries
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt
import seaborn as sns

# ------------------- Step 1: Load and Clean Data ------------------- #
# Load the dataset into a DataFrame
df = pd.read_csv("../Projects/GrennTree/test.csv")

# Combine 'Subject' and 'body' columns into a single 'full_text' column
df["full_text"] = df["Subject"].fillna("") + " " + df["body"].fillna("")

# Rename the 'from' column to 'Employee' for clarity
df.rename(columns={"from": "Employee"}, inplace=True)

# Convert the 'date' column to datetime format and drop rows with invalid dates
df["date"] = pd.to_datetime(df["date"], errors='coerce')
df = df.dropna(subset=["date", "full_text"])

# ------------------- Step 2: Sentiment Scoring ------------------- #
# Initialize SentimentIntensityAnalyzer for sentiment analysis
analyzer = SentimentIntensityAnalyzer()

# Calculate the sentiment compound score for each 'full_text'
df["Sentiment_Compound"] = df["full_text"].apply(lambda x: analyzer.polarity_scores(x)["compound"])

# ------------------- Step 3: Text Vectorization (TF-IDF) ------------------- #
# Apply TF-IDF vectorization on the 'full_text' column, limiting to 500 features
tfidf = TfidfVectorizer(max_features=500)  # limit to 500 features
X_tfidf = tfidf.fit_transform(df["full_text"])

# ------------------- Step 4: Target Variable ------------------- #
# Set the target variable as the Sentiment Compound scores
y = df["Sentiment_Compound"]

# ------------------- Step 5: Train-Test Split with Index Tracking ------------------- #
# Track original indices of the DataFrame before splitting into train-test sets
original_indices = df.index

# Split the data into training and testing sets (80-20 split)
X_train, X_test, y_train, y_test, idx_train, idx_test = train_test_split(
    X_tfidf, y, original_indices, test_size=0.2, random_state=42
)

# ------------------- Step 6: Train the Model ------------------- #
# Initialize and train a Random Forest Regressor model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict sentiment scores on the test set
y_pred = model.predict(X_test)

# ------------------- Step 7: Evaluation ------------------- #
# Calculate Mean Squared Error (MSE) and R score to evaluate model performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Output the evaluation metrics
print(f" Mean Squared Error: {mse:.4f}")
print(f" R Score: {r2:.4f}")

# ------------------- Step 8: Visualization ------------------- #
# Plot the Actual vs Predicted Sentiment scores
plt.figure(figsize=(8, 6))
sns.regplot(x=y_test, y=y_pred, scatter_kws={'alpha':0.5}, line_kws={"color": "red"})
plt.xlabel("Actual Sentiment Scores")
plt.ylabel("Predicted Sentiment Scores")
plt.title("Actual vs Predicted Sentiment (Using TF-IDF Features)")
plt.grid(True)
plt.tight_layout()
plt.show()

# ------------------- Step 9: Export Results with Sender Info ------------------- #
# Create a DataFrame to store the test set results along with sender info and predictions
results_df = pd.DataFrame({
    "Employee": df.loc[idx_test, "Employee"].values,
    "Original_Text": df.loc[idx_test, "full_text"].values,
    "Actual_Sentiment": y_test.values,
    "Predicted_Sentiment": y_pred
})

# Export the results to an Excel file
results_df.to_excel("sentiment_predictions.xlsx", index=False)

# Display the first 10 rows of the results
results_df.head(10)
```
